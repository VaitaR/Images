```python
%matplotlib inline
from PIL import Image, ImageDraw
import numpy as np
from skimage.io import imread, imshow, imsave
import numpy as np
from scipy import misc
import imageio
import matplotlib.pyplot as plt #построение графиков
```

```python
im = imageio.imread('1.jpg') #Считываем изображение
width = im.shape[0] # Определяем ширину
height = im.shape[1] # Определяем высоту
#imshow(im[:, :, :]) # Выгружаем изображение
#im[:,:] # Выгружаем значения пикселей
Intensity_values = [] # Пустой список, куда пойдут значения интенсивностей красного канала
for i in range(width):
    for j in range(height):
        Intensity_values.append(im[i, j][0]) # Вставляем в список значения интенсивностей красного канала
summation = 0 # Переменная, куда буду складывать разности каждого элемента со средним
n = len(Intensity_values) # Длина списка
average_mean = sum(Intensity_values) / len(Intensity_values) # Вычисляю среднее значение элемента списка
for y in range (0,n): #
    difference = Intensity_values[y] - average_mean #Вычисляю разность для каждого элемента списка
    squared_difference = difference ** 2
    summation = summation + squared_difference
MSE = summation / n
sigma = MSE ** 0.5
print (sigma)
            
```
